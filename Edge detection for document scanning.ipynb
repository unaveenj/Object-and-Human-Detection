```python
import cv2
import numpy as np
```

```python
doc = cv2.imread("note_detection.jpg")


#Step 1 Resize the image to desired size
doc = cv2.resize(doc,(600,400))
original = doc.copy()







```

```python
#Step2 Convert the image to greyscale
doc = cv2.cvtColor(doc,cv2.COLOR_BGR2GRAY)
# cv2.imshow("Grayscale",doc)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
#Step3 Apply Gaussian blur for better edge detection
doc = cv2.GaussianBlur(doc,(5,5),0)
# cv2.imshow("GuassianBlur",doc)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
#Step 4 Apply canny filter
canny = cv2.Canny(doc,100,120)
# cv2.imshow("Canny",canny)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
#Step5 Drawing countour boundary
contours,heriarchy = cv2.findContours(canny,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

print("Countours found are : " +str(len(contours)))

# Draw all contours
# Use '-1' as the 3rd parameter to draw all
cv2.drawContours(original, contours, -1, (255,0,0), 3)
cv2.imshow('Contours', original)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Corner detection using Good features to track is a better corner detection algorithm
import cv2
import numpy as np
img = cv2.imread('note_detection.jpg')
img= cv2.resize(img,(800,600))
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# Specify top 10 corners
corners = cv2.goodFeaturesToTrack(gray, 10, 0.01, 150)

for corner in corners:
    x, y = corner[0]
    x = int(x)
    y = int(y)
    cv2.rectangle(img,(x-10,y-10),(x+10,y+10),(0,255,0), 2)
    
cv2.imshow("Corners Found", img)
cv2.waitKey()
cv2.destroyAllWindows()
```
